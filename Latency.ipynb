## "Time Between"
### Simple tap based system to access the latency between visual and sound occurrences
This was made to find out how much time is spent between taps. I made this because I always get lag when trying to watch a video through bluetooth speakers.

```python
import numpy as np
import time
import IPython.display as ipd
```

```python
#Basic Configuration

amnt = input("Set the number of cycles for acessing the latency: ")

try:
    int(amnt)
except ValueError: 
    print ("Not an integer.")
    
amnt = int(amnt)
```

```python
#creating a numpy matrix to store the values
delay0 = np.zeros((amnt,1))
delay1 = np.zeros((amnt,1))
print ("Created two matrixes of", delay0.shape[0], "lines and", delay1.shape[1], "column.")
```

```python
#creating a sine wave
#obtained from https://musicinformationretrieval.com/ipython_audio.html

sr = 44100 # sample rate
T = 1    # seconds
t = np.linspace(0, T, int(T*sr), endpoint=False) # time variable
x = 0.5*np.sin(2*np.pi*440*t)                # pure sine wave at 440 Hz


snd=ipd.Audio(x, rate=sr, autoplay=True) # load a NumPy array
```

```python
print ("Press ENTER when you hear a sound in...")
time.sleep(1)
print ("3...")
time.sleep(1)
print ("2...")
time.sleep(1)
print ("1...")
for i in range (amnt):
    time.sleep(1)
    delay0 [i] = time.time()
    display(snd)
    s = input("Press ENTER when you hear a sound")
    delay1 [i] = time.time()
    if i != amnt-1:
        print ("OK! Again.")
        time.sleep(1)
        print ("2...")
        time.sleep(1)
        print ("1...")
    
print ("Done.")
```

```python
avgdelay = np.subtract(delay0, delay1)
avgdelay = np.average(avgdelay)
avgdelay = np.multiply(avgdelay,-1000)
avgdelay = np.round(avgdelay, decimals=3)

print ("Average delay:", avgdelay, "milliseconds")
```
